```python
from transformers import pipeline

# Load the sentiment analysis pipeline with a pre-trained DistilBERT model
# "distilbert-base-uncased-finetuned-sst-2-english" is a model specifically trained for sentiment analysis
classifier = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english")

# Test with some example sentences
print(classifier("I love learning about Small Language Models! This is so cool."))
print(classifier("This movie was terrible. What a waste of time."))
print(classifier("The weather today is neither good nor bad."))
print(classifier("This product is alright, but it could be better."))
```

```python
from transformers import pipeline, set_seed

# Load the text generation pipeline with DistilGPT2
# "distilgpt2" is a smaller, faster version of GPT-2
generator = pipeline("text-generation", model="distilgpt2")

# Set a seed for reproducibility (so you get similar results each time)
set_seed(42)

# Generate text
print("--- Generated Text 1 ---")
print(generator("Once upon a time, in a land far, far away,", max_length=50, num_return_sequences=1))

print("\n--- Generated Text 2 ---")
print(generator("The secret to success is", max_length=50, num_return_sequences=1))

print("\n--- Generated Text 3 (more creative) ---")
print(generator("If robots could feel emotions, they would", max_length=70, num_return_sequences=1, temperature=0.9)) # Higher temperature for more creativity
```
